```python
import numpy as np
import matplotlib.pyplot as plt

from hackrl.utils.dataset_scores import get_dataset_scores
```

```python
dbfilename = "/home/bartek/Workspace/CW/dungeonsdata-neurips2022/experiment_code/ttyrecs/ttyrecs.db"
```

```python
dataset_scores = get_dataset_scores("altorg", dbfilename)
alt_scores = np.array(list(dataset_scores.values()))
plt.hist(alt_scores, bins=200)
plt.axvline(x=alt_scores.mean(), c="red", alpha=0.3)
plt.ylabel('Number of games (log scale)')
plt.xlabel('scores')
plt.yscale('log')
plt.title(f"NLD-NAO (mean={alt_scores.mean().astype(int)})")
```
